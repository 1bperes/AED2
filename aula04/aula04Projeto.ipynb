```python
def buscar_livro(biblioteca, titulo_procurado):
    inicio = 0 
    fim = len(biblioteca) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2 
        livro_atual = biblioteca[meio]

        if livro_atual['Titulo'] == titulo_procurado:
            return livro_atual
        elif livro_atual['Titulo'] < titulo_procurado:
            inicio = meio + 1
        else:
            fim = meio - 1

    return None

meus_livros = [
    {"Titulo": "1984", "Autor": "George Orwell", "Ano": 1949}, 
    {"Titulo": "A Metamorfose", "Autor": "Franz Kafka", "Ano": 1915},
    {"Titulo": "Dom Casmurro", "Autor": "Machado de Assis", "Ano": 1899},
    {"Titulo": "O Rei de Amarelo", "Autor": "Robert W. Chambers", "Ano": 1895}
]

resultado = buscar_livro(meus_livros, "O Rei de Amarelo")

if resultado:
    print(f"Encontrado: {resultado['Titulo']} de {resultado['Autor']}")
else:
    print("Livro não encontrado")
```
