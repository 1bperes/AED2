```python
def buscar_produto(catalogo, nome_procurado):
    inicio = 0
    fim = len(catalogo) - 1

    while inicio <= fim:
        meio = (inicio + fim) // 2
        produto_atual = catalogo[meio]

        if produto_atual['nome'] == nome_procurado:
            return produto_atual
        elif produto_atual['nome'] < nome_procurado:
            inicio = meio + 1
        else:
            fim = meio - 1

    return None

catalogo_produtos = [
    {"nome": "Adaptador USB-C", "preco": 45.00, "descricao": "Conector multifuncional de alta velocidade."},
    {"nome": "Cafeteira Expressa", "preco": 450.00, "descricao": "Cafeteira automática com moedor integrado."},
    {"nome": "Fone Bluetooth", "preco": 299.90, "descricao": "Fone com cancelamento de ruído e bateria de 20h."},
    {"nome": "Luminária de Mesa", "preco": 85.00, "descricao": "Luminária LED com ajuste de intensidade."},
    {"nome": "Mochila Impermeável", "preco": 180.00, "descricao": "Mochila resistente com compartimento para notebook."},
    {"nome": "Relógio Digital", "preco": 120.00, "descricao": "Relógio à prova d'água com cronômetro."},
    {"nome": "Smartphone 5G", "preco": 2100.00, "descricao": "Tela de 6.7 polegadas e 128GB de armazenamento."}
]



resultado = buscar_produto(catalogo_produtos, "Mochila Impermeável")

if resultado:
    print(f"Produto: {resultado['nome']}")
    print(f"Preço: R$ {resultado['preco']:.2f}")
    print(f"Descrição: {resultado['descricao']}")
else:
    print("Produto não localizado no catálogo.")
```
